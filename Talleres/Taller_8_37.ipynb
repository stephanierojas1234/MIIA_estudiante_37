### Ejercicio 4
Utiliza `ipywidgets` para filtrar las observaciones de la base de datos de acuerdo con los siguentes criterios:
1. Nivel máximo de escolaridad alcanzado.
2. Número de niños en el hogar.
3. Número de adolescentes en el hogar.

```python
# YOUR CODE HERE

from ipywidgets import interactive

mercadeo["Kidhome"] = mercadeo["Kidhome"].astype(str)
mercadeo["Teenhome"] = mercadeo["Teenhome"].astype(str)

def valores_filtro_columna(columna):
    columna = columna.astype(str)       # Convertimos en str los valores de la columna que entra por parámetro.
    valores = columna.unique().tolist() # Guardamos una lista con los valores únicos de la columna que entra por parámetro.
    valores = sorted(valores)           # Organizamos la lista de valores únicos de manera ascedente.
    valores.insert(0, "Todos")          # Agregamos el elemento "Todos" en la primera posición de la lista.
    return valores 

def filtrar(Escolaridad, Ninhos, Adolecentes):
    # Guardamos las columnas de interés en un nuevo DataFrame.
    # Los siguientes cuatro casos ajustan los registros a mostrar de acuerdo con la selección del usuario.
    if  Escolaridad != "Todos" and Ninhos != "Todos" and Adolecentes!= "Todos":
        display(mercadeo[(mercadeo["Education"] == Escolaridad) & (mercadeo["Kidhome"] == Ninhos)
                   & (mercadeo["Teenhome"] == Adolecentes)])
    elif Escolaridad == "Todos" and Ninhos != "Todos" and Adolecentes!= "Todos":
         display(mercadeo[(mercadeo["Kidhome"] == Ninhos)
                   & (mercadeo["Teenhome"] == Adolecentes)])
    elif Escolaridad != "Todos" and Ninhos == "Todos" and Adolecentes!= "Todos":
         display(mercadeo[(mercadeo["Education"] == Escolaridad)
                   & (mercadeo["Teenhome"] == Adolecentes)])
    elif Escolaridad != "Todos" and Ninhos != "Todos" and Adolecentes == "Todos":
         display(mercadeo[(mercadeo["Education"] == Escolaridad)
                   & (mercadeo["Kidhome"] == Ninhos)])
    elif Escolaridad != "Todos" and Ninhos == "Todos" and Adolecentes == "Todos":
         display(mercadeo[(mercadeo["Education"] == Escolaridad)])
    elif Escolaridad == "Todos" and Ninhos != "Todos" and Adolecentes == "Todos":
         display(mercadeo[(mercadeo["Kidhome"] == Ninhos)])
    elif Escolaridad == "Todos" and Ninhos == "Todos" and Adolecentes != "Todos":
         display(mercadeo[(mercadeo["Teenhome"] == Adolecentes)])
    else:
        display(mercadeo)
        
interactive_plot = interactive(
                                filtrar, 
                                Escolaridad = valores_filtro_columna(mercadeo["Education"]),
                                Ninhos = valores_filtro_columna(mercadeo["Kidhome"]),
                                Adolecentes = valores_filtro_columna(mercadeo["Teenhome"]),
                              )
display(interactive_plot)


#raise NotImplementedError()
```

### Ejercicio 5
Utiliza `ipywidgets` y `plotly` para generar un gráfico con controles que:
1. Cuente con un control para seleccionar una de las variable de Gastos frente a los ingresos.
1. Genere un gráfico de dispersión
2. Genere un diagrama de caja.

```python
# YOUR CODE HERE

#Diagrama de dispersion
lista_gastos=list(mercadeo.columns.values)[9:15]
menu_gastos = widgets.Dropdown(options = lista_gastos)

fig1 = go.FigureWidget(
                        data = go.Scatter(mode ='markers'),  
    
                        layout = go.Layout(
                                            title = 'Ingresos vs Gastos',
                                            xaxis_title = 'Ingresos',
                                            yaxis_title = 'Gastos',
                                          )
                     )

fig2 = go.FigureWidget(
                        data = go.Scatter(mode ='markers'),  
    
                        layout = go.Layout(
                                            title = 'Ingresos vs Gastos',
                                            xaxis_title = 'Ingresos',
                                            yaxis_title = 'Gastos',
                                          )
                     )

def filtrar():  
    fig1.data[0].x = mercadeo['Income'] # Establecemos los datos en el eje x.
    fig1.data[0].y=mercadeo[menu_gastos.value]
    fig2.data[0].x = mercadeo['Income'] # Establecemos los datos en el eje x.
    fig2.data[0].y=mercadeo[menu_gastos.value]
    
def respuesta(change):
    filtrar()
    
menu_gastos.observe(respuesta, names = 'value')

filtrar()

parte_inferior = widgets.HBox([menu_gastos])

visualizacion = widgets.VBox(
              [parte_inferior,
              fig1,
              fig2]
            )

display(visualizacion)



#raise NotImplementedError()
```
